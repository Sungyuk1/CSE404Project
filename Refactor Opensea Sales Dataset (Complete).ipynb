The purpose of this Juptyer Notebook is to refactor the dataset being used to have a NFT name and then a list of wallet addresses that have purchased the NFT collection between 2019 and 2021

The data set we are using is called OpenSea NFT Sales 2019 - 2021. It can be found with this link : https://www.kaggle.com/datasets/bryanw26/opensea-nft-sales-2019-2021

```python
#importing neccessary libraries
import numpy as np
import pandas as pd
```

```python
#Load in the dataset
data_frame = pd.read_csv('OpenseaData/OpenSea_NFT_Sales_2019_2021.csv', error_bad_lines=False)
```

```python
data_frame.head()
```

```python
#Remove NaN rows
data_frame = data_frame.dropna()
```

```python
data_frame.describe()
```

```python
data_frame.head()
```

```python
#Drop the columns we do not care about. We only care about collection name and winner_account.address, but others might be useful later
# Note : axis=1 means we are dropping the column, not the row.
data_frame = data_frame.drop('sales_datetime', axis=1)
data_frame = data_frame.drop('asset.id', axis=1)
data_frame = data_frame.drop('id', axis=1)
data_frame = data_frame.drop('asset.collection.short_description', axis=1)
data_frame = data_frame.drop('asset.permalink', axis=1)
data_frame = data_frame.drop('seller.user.username', axis=1)
data_frame = data_frame.drop('Category', axis=1)
```

```python
data_frame.head()
```

```python
data_frame_minimal = data_frame.drop('asset.name', axis=1)
data_frame_minimal = data_frame_minimal.drop('total_price', axis=1)
data_frame_minimal = data_frame_minimal.drop('payment_token.name', axis=1)
data_frame_minimal = data_frame_minimal.drop('payment_token.usd_price', axis=1)
data_frame_minimal = data_frame_minimal.drop('asset.num_sales', axis=1)
data_frame_minimal = data_frame_minimal.drop('seller.address', axis=1)
```

```python
data_frame_minimal.head()
```

```python
#Pandas dataframe.groupby() function is used to split the data into groups based on some criteria
data_frame_right_format = data_frame_minimal.groupby('asset.collection.name')['winner_account.address'].apply(list)
```

```python
data_frame_right_format.head()
```

```python
#Note : toexcel writes object to an Excel Sheet
data_frame_right_format.to_excel('OpenSeaSalesDataRightFormat.xlsx')
```

```python
data_frame_right_format.to_csv('OpenSeaSalesDataRightFormat_csv.csv')
```

```python
address_to_collections_data_frame = data_frame_minimal.groupby('winner_account.address')['asset.collection.name'].apply(list)
```
